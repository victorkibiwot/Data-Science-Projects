# Bankruptcy prediction using Python


Importing required packages. We shall be using Logistic Regression to make the classifier

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = pd.read_csv('data.csv')
```

```python
data.head()
```

Lets check some correlation before performing any training

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.heatmap(data.corr())
plt.show()
```

```python
X = data.drop(['Bankrupt?'], axis = 'columns')
y = data['Bankrupt?'] 
```

```python
#Lets prepare the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
predictions = log_reg.predict(X_test)
accuracy = accuracy_score(y_test, predictions) 

print("Accuracy obtained is: " + str(accuracy.round(4)*100) + "%")
```

We can see that the LogisticRegression classifier works pretty well with an accuracy of approximately 96%.
