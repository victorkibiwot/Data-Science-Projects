```python
import pandas as pd

df = pd.read_csv("bike_data.csv", index_col= 'Date', parse_dates=True)
df.head()
```

```python
df.columns = ["Total", "East", "West"]
df.head()
```

```python
df.dropna().describe()
```

Some data vizualization is important. Lets plot some data

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
df.plot()
plt.ylabel("Hourly bicycle count")
plt.show()
```

Since its too hard to visualize, we will resample the data on a weekly basis.

```python
weekly = df.resample('W').sum()
weekly.plot(style=[':','--','-'])
            
plt.ylabel('Weekly Bicycle count')
plt.show()
```

This shows us some interesting seasonal trends. More people cycle during the summer than the winter

Another way of resampling the data is using the pd.rolling_mean() function.

Lets try a 30 day rolling mean of our data.

Note that performing the rolling mean retains the same sample size

```python
daily = df.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':','--','-'])
plt.ylabel("Mean daily count")
plt.show()
```

he jaggedness of the result is due to the hard cutoff of the window. We can get a smoother version of a rolling mean using a window functionfor example, a Gaussian window.

```python
daily.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':','--', '-'])
plt.show()
```

### Lets dig deeper into the data 

For instance lets explore the average traffic as a function of the time of day.
 
We can use groupby() function to achieve this.

```python
import numpy as np
by_time = df.groupby(df.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks= hourly_ticks, style=[':', '--', '-'])
plt.ylabel("Traffic according to time")
plt.show()
```
